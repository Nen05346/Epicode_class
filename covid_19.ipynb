The function below maages request n°3 in a modular way 

```python
def statistic_cmp(df, attr, cmp_attr, cmp_1, cmp_2):
    """
    Max, mean, percentage.
    The function accepts 5 parameters:
        - df: a pandas dataframe
        - attr: a string indicating tha attribute we want to compare
        - cmp_attr: a string indicating the attribute of the dataframe on which we want to calculate the statistics
        - cmp_1: one of the unique values of the attribute attr
        - cmp_2: one of the unique values of the attribute attr
        """
    # cmp_1 
    max1 = df[df[attr] == cmp_1].describe().loc['max'][cmp_attr]
    mean1 = df[df[attr] == cmp_1].describe().loc['mean'][cmp_attr]
    perc1 = ((df[df[attr] == cmp_1][cmp_attr].sum()) / (df[cmp_attr].sum())) * 100
    # cmp_2
    max2 = df[df[attr] == cmp_2].describe().loc['max'][cmp_attr]
    mean2 = df[df[attr] == cmp_2].describe().loc['mean'][cmp_attr]
    perc2 = ((df[df['continent'] == cmp_2][cmp_attr].sum()) / (df[cmp_attr].sum())) * 100
    
    stats = {
        'descriptors': ['max', 'mean', '%'],
        cmp_1: [max1, mean1, perc1],
        cmp_2: [max2, mean2, perc2],
    }
    
    statistics = pd.DataFrame(stats)
    
    return statistics

         
statistic_cmp(df, 'continent', 'total_cases', 'Europe', 'South America')
```
